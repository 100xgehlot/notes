```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
```

```python
dataset = pd.read_csv("dataset/Mall_Customers.csv")
x = dataset.iloc[:,[3,4]].values
x
```

```python
import scipy.cluster.hierarchy as sch
dendrogram =  sch.dendrogram(sch.linkage(x,method="ward"))
plt.title("Dendrogram")
plt.xlabel("Customers")
plt.ylabel("Euclidean Distance")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=5,affinity="euclidean",linkage = "ward" )
y_hc = hc.fit_predict(x)
```

```python

plt.scatter(x[y_hc == 0,0], x[y_hc == 0,1], s=300, c='red', label='Clustering 1')
plt.scatter(x[y_hc == 1,0], x[y_hc == 1,1], s=300, c='blue', label='Clustering 2')
plt.scatter(x[y_hc == 2,0], x[y_hc == 2,1], s=300, c='green', label='Clustering 3')
plt.scatter(x[y_hc == 3,0], x[y_hc == 3,1], s=300, c='orange', label='Clustering 4')
plt.scatter(x[y_hc == 4,0], x[y_hc == 4,1], s=300, c='black', label='Clustering 5')

plt.title("Agglomerative Clustering")
plt.xlabel("Annual income")
plt.ylabel("Spending Score")
plt.legend()
plt.show()
```
