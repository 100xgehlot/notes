```python
# plot contor plot
import numpy as np
import matplotlib.pyplot as plt
```

```python
#generate r and theta array
rad_arr  = np.radians(np.linspace(0,360,20))
r_arr = np.arange(0,1,.1)
```

```python
#define function
def func(r,theta):
    return r*np.sin(theta)
r,theta = np.meshgrid(r_arr,rad_arr)
```

```python
#get the values of response variable 
values = func(r,theta)
```

```python
#plot the polar coordinates
fig,ax = plt.subplots(subplot_kw=dict(projection="polar"))
ax.contourf(theta,r,values,cmp="Spectural_r")
plt.show()
```
